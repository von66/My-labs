```python
from mlxtend.frequent_patterns import apriori
```

```python
dataset = [['Drink', 'Nuts', 'Diaper'],

      ['Drink', 'Coffee', 'Diaper'],

      ['Drink', 'Diaper', 'Eggs'],

      ['Nuts', 'Eggs', 'Milk'],

      ['Nuts', 'Coffee', 'Diaper', 'Eggs', 'Milk'] ,
          ['Milk', 'orange', 'tea']]

dataset
```

# 1.2 

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
TranEncod = TransactionEncoder()
te_ary = TranEncod.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=TranEncod.columns_)
```

```python
df
```

```python
apriori(df, min_support=0.2, use_colnames=True)
```

# 1.3 

```python
frequent_itemsets = apriori(df, min_support=0.20, use_colnames=True)

frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &

          (frequent_itemsets['support'] >= 0.20) ]

```

```python
frequent_itemsets[ frequent_itemsets['itemsets'] == {'Diaper', 'Drink'} ]
```

```python
dataset = [['l1', 'l2', 'l5'],

      ['l2', 'l4'],

      ['l2', 'l3'],

      ['l1', 'l2', 'l4'],
           ['l1', 'l3'],
           ['l2', 'l3'],
           ['l2', 'l3'],

      ['l1', 'l2', 'l3', 'l5'] ,
          ['l1', 'l2', 'l3']]

dataset
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
TranEncod = TransactionEncoder()
te_ary = TranEncod.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=TranEncod.columns_)
```

```python
df
```

```python
frequent_itemsets = apriori(df, min_support=0.20, use_colnames=True)

frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &

          (frequent_itemsets['support'] >= 0.20) ]
```
